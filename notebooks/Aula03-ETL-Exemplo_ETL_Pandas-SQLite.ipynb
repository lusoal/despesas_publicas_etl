```python
#descobre o encoding do arquivo
import chardet
def find_encoding(fname):
    r_file = open(fname, 'rb').read()
    result = chardet.detect(r_file)
    charenc = result['encoding']
    return charenc
```

```python
import sqlite3
from sqlite3 import Error
 
connection = None
try:
    connection = sqlite3.connect(".\\orgao_db.sqlite")
    print("Conectado no SQLite com sucesso")
    
    drop_orgao_table = """
    DROP TABLE IF EXISTS orgao;
    """
    cursor = connection.cursor()
    cursor.execute(drop_orgao_table)
    print("Tabela apagada.")

    create_orgao_table = """
    CREATE TABLE orgao (
      id_orgao INTEGER PRIMARY KEY,
      nm_orgao TEXT NOT NULL
    );
    """
    cursor = connection.cursor()
    cursor.execute(create_orgao_table)
    print("Tabela criada.")
    
    query = "INSERT INTO orgao (id_orgao, nm_orgao) VALUES (?, ?)"
    
    cursor.executemany(query,values)
    connection.commit()
    print(cursor.rowcount, "linha(s) inserida(s)")
    
except Error as e:
    connection.rollback()
    print(f"O erro '{e}' ocorreu")
```

```python
cursor = connection.cursor()

## definindo a consulta
query = "SELECT * FROM orgao"

## pegando os registros na tabela
cursor.execute(query)

## buscando todos os registros do objeto cursos; eles vem em uma lista
records = cursor.fetchall()
print(type(records))

for record in records:
    print(record)
```
