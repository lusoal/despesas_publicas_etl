```python
# DAO layer of script

def create_connection(host_name, user_name, user_password, db_name):
    connection = None
    try:
        connection = mysql.connector.connect(
            host=host_name,
            user=user_name,
            passwd=user_password,
            database=db_name
        )
        print("Connection to MySQL DB successful")
    except Error as e:
        print(f"The error '{e}' occurred")

    return connection

def execute_read_query(connection, query):
    cursor = connection.cursor()
    result = None
    try:
        cursor.execute(query)
        result = cursor.fetchall()
        return result
    except Error as e:
        print(f"The error '{e}' occurred")
        
def execute_query(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        connection.commit()
        print("Query executed successfully")
        
        # imprimir o id criado pelo autoincrement somente nos inserts, 
        # pois nos demais comandos dá resultados incoerentes
        if query.lstrip().upper().startswith('INSERT'):
            print('Last insert id:', cursor.lastrowid)
    except Error as e:
        print(f"The error '{e}' occurred")
```

Acima estao as funcoes utilizadas para se comunicar com o DB

```python
# Dataframe Functions

def upper_column_head(df, columns):
    df_upperized = df
    for columns in columns:
        df_upperized.rename(columns={columns: columns.upper()}, inplace=True)
    return df_upperized


def merge_dataframes(df_orcamento, df_exec_mes):
    result = pd.merge(df_orcamento,
             df_exec_mes[['NOME ÓRGÃO SUPERIOR', 'NOME ÓRGÃO SUBORDINADO', 
                                'NOME UNIDADE ORÇAMENTÁRIA', 'NOME FUNÇÃO', 'NOME SUBFUNÇÃO',
                               'NOME PROGRAMA ORÇAMENTÁRIO', 'NOME AÇÃO', 'VALOR LIQUIDADO (R$)']],
             on=['NOME ÓRGÃO SUPERIOR', 'NOME ÓRGÃO SUBORDINADO', 
                                'NOME UNIDADE ORÇAMENTÁRIA', 'NOME FUNÇÃO', 'NOME SUBFUNÇÃO',
                               'NOME PROGRAMA ORÇAMENTÁRIO', 'NOME AÇÃO'])
    result['ORÇAMENTO INICIAL (R$)'] = result['ORÇAMENTO INICIAL (R$)'].div(12)
    return result

def group_dataframes(dataframe, group_values, convert_value):
    dataframe = dataframe[group_values]
    dataframe[convert_value] = (dataframe[convert_value].replace(",", ".", regex = True).astype(float)).round(2)
    dataframe = dataframe.groupby(group_values).sum()
    dataframe = dataframe.reset_index()
    return dataframe
```

Acima estao as funcoes em comum utilizadas para manipular dataframes

```python

df_orcamento_grouped = df_orcamento.groupby(['NOME ÓRGÃO SUPERIOR', 'NOME ÓRGÃO SUBORDINADO', 'NOME UNIDADE ORÇAMENTÁRIA']).mean()
df_orcamento_grouped = df_orcamento_grouped.reset_index()
```

```python
QUERY  = "INSERT INTO dimensao_orgao(orgao_superior, orgao_subordinado, unidade_orcamentaria)"

for org_sup, org_sub, unid_orc in zip(df_orcamento_grouped['NOME ÓRGÃO SUPERIOR'],
                df_orcamento_grouped['NOME ÓRGÃO SUBORDINADO'], df_orcamento_grouped['NOME UNIDADE ORÇAMENTÁRIA']):
    query_for = f"{QUERY} VALUES('{org_sup}', '{org_sub}', '{unid_orc}')"
    execute_query(conn, query_for)
```

```python
# Transformando Header em maiusculo

execucao_columns = list(df_execucao_jan.columns)
df_execucao_jan = upper_column_head(df_execucao_jan, execucao_columns)

df_execucao_jan.columns
```

```python
grouped_df_orcamento = group_dataframes(df_orcamento, ['NOME ÓRGÃO SUPERIOR', 'NOME ÓRGÃO SUBORDINADO', 'NOME UNIDADE ORÇAMENTÁRIA',
                            'NOME PROGRAMA ORÇAMENTÁRIO','NOME FUNÇÃO', 'NOME SUBFUNÇÃO', 'NOME AÇÃO',
                            'ORÇAMENTO INICIAL (R$)'], 'ORÇAMENTO INICIAL (R$)')
```

```python
grouped_df_execucao_jan = group_dataframes(df_execucao_jan, ['NOME ÓRGÃO SUPERIOR', 'NOME ÓRGÃO SUBORDINADO', 'NOME UNIDADE ORÇAMENTÁRIA',
                            'NOME PROGRAMA ORÇAMENTÁRIO','NOME FUNÇÃO', 'NOME SUBFUNÇÃO', 'NOME AÇÃO',
                            'VALOR LIQUIDADO (R$)'], 'VALOR LIQUIDADO (R$)')
```

```python
md = merge_dataframes(grouped_df_orcamento, grouped_df_execucao_jan)
md.head(2)
```

```python
mes = ["1"]
ano = "2019"


for org_sup, org_sub, unid_orc, prog_orc, func, subfunc, acao, inicial, liquidado in zip(md['NOME ÓRGÃO SUPERIOR'],
                                                                                        md['NOME ÓRGÃO SUBORDINADO'],
                                                                                        md['NOME UNIDADE ORÇAMENTÁRIA'],
                                                                                        md['NOME PROGRAMA ORÇAMENTÁRIO'],
                                                                                        md['NOME FUNÇÃO'],
                                                                                        md['NOME SUBFUNÇÃO'],
                                                                                        md['NOME AÇÃO'],
                                                                                        md['ORÇAMENTO INICIAL (R$)'],
                                                                                        md['VALOR LIQUIDADO (R$)']):
    
    orgao_query = f"SELECT id_orgao FROM dimensao_orgao WHERE orgao_superior='{org_sup}' and orgao_subordinado='{org_sub}' and unidade_orcamentaria='{unid_orc}'"
    id_orgao = execute_read_query(conn, orgao_query)[0][0]
    
    programa_query = f"SELECT id_dimensao_programa FROM dimensao_programa WHERE programa_orcamentario='{prog_orc}' and acao='{acao}'"
    id_programa = execute_read_query(conn, programa_query)[0][0]
    
    area_atuacao_query = f"SELECT id_dimensao_area_atuacao FROM dimensao_area_atuacao WHERE funcao='{func}' and subfuncao='{subfunc}'"
    id_area_atuacao = execute_read_query(conn, area_atuacao_query)[0][0]
    
    tempo_query = f"SELECT id_dimensao_temporal FROM dimensao_temporal WHERE mes='{mes[0]}' and ano='{ano}'"
    id_tempo = execute_read_query(conn, tempo_query)[0][0]
    
    
    insert_query = f"""INSERT INTO fato_orcado_liquidado(valor_orcado, valor_liquidado, 
    dimensao_orgao_id_orgao, dimensao_programa_id_dimensao_programa, dimensao_area_atuacao_id_dimensao_area_atuacao, dimensao_temporal_id_dimensao_temporal) VALUES({inicial}, {liquidado}, {id_orgao}, 
    {id_programa}, {id_area_atuacao}, {id_tempo})"""
    
    execute_query(conn, insert_query)
```
